# Are golfers focused on the right metrics?

## <b>Background</b>
Most golf coaches encourage players to focus on three metrics to improve their golf score (a lower score is better in golf): 
- <b>Greens in regulation (GIR):</b> The percent of time a player was able to hit the green in regulation (greens hit in regulation/holes played)
- <b>Driving accuracy:</b> The percentage of time a tee shot comes to rest in the fairway (regardless of club)
- <b>Scrambling:</b> The percent of time a player misses the green in regulation, but still makes par or better

After spending thousands of hours practicing golf in high school and college, <b>I wanted to know if there was any statistical evidence that proved these metrics significantly impacted a player's score.</b>

## Data set
I chose PGA Tour ShotLink data from 1980 to 2019 (www.shotlink.com and www.pgatour.com).  This is the largest set of data available on these metrics.  The data is manually collected by volunteers who follow every PGA Tour player and record every shot made in every PGA Tour tournament.  I was looking into scraping the data when I realized someone on Kaggle had already done this (https://www.kaggle.com/bradklassen/pga-tour-20102018-data).  I instead used this data set, formatted it, and cleaned out obvious errors.


## Experiment #1: Does a player's <u>greens in regulation (GIR) percentage</u> significantly impact their Score?
- <b>Method:</b> Linear regression. Is there a significant linear relationship between "Score" and "GIR"?
- <b>Null hypothesis (H0):</b> There is <u>no</u> association between Score and GIR (Beta1 (aka slope) = 0)
- <b>Alternate hypothesis (HA):</b> There <u>is</u> an association between Score and GIR (Beta1 is less than or greater than 0)
- <b>Sample size:</b> 124,540 PGA Tour players' average % of greens made in regulation per tournament (typically 72 holes per tournament)
- <b>Alpha:</b> 0.05

<b><i>Takeaway: The association between GIR% and Score <u>is</u> statistically significant</i></b>

- <b>P-value:</b> 0.000. The P-value of nearly zero is much smaller than 5%, so we can reject the null hypothesis. There is a statistically significant difference between the two sample means. There is strong evidence that there is a real association between a player's score and the percent of greens in regulation they make in this population. Therefore, GIR is a predictor of Score.

<b><i>Takeaway: There is weak correlation between the GIR% and Score </i></b>
- <b>Pearson Correlation Coefficient:</b> -0.576033.
- <b>R-squared:</b> 0.332. This would classify as "weak" correlation based on Evans or "medium" correlation based on Cohen.


<b><i>Takeaway: With 95% confidence, a 1% increase in GIR% is estimated to lower the population mean's Score by 12.7% to 12.9%</i></b>
- <b>95% confidence interval:</b> 12.7% - 12.9%

<b><i>Takeaway: The shape of the data and residuals appear to be normally distributed</i></b>
- Based on charts later in this analysis:
- <b>Scatterplot:</b> The shape of the data is relatively symmetric
- <b>Residual plot, normal probability plot (Q-Q), histogram of residuals:</b> The data appears normally distributed.

## Experiment #3: Does a player's <u>scrambling percentage</u> significantly impact their score?

- <b>Alternate hypothesis (HA):</b> There <u>is</u> an association between Score and Scrambling (Beta1 is less than or greater than 0)

<b><i>Takeaway: There is weak correlation between the Scrambling % and Score </i></b>

<b><i>Takeaway: With 95% confidence, a 1% increase in Scrambling percentage is estimated to lower the population mean's Score by 7.6% to 7.8%</i></b>
- <b>95% confidence interval:</b> 7.6% - 7.8%

## Conclusion

- The association <b><i>is statistically significant</i></b> between each of the three metrics and Score
- Correlation between the three metrics and Score was <b><i>weak</i></b> (GIR, Scrambling) to <b><i>very weak</i></b> (Driving Accuracy)
- With 95% confidence, <b><i>a 1% improvement in GIR is estimated to lower the population mean's score by 12.7% to 12.9% (8.97 to 9.11 strokes)</i></b>, which is nearly 2x the amount for Scrambling and over 5x the amount for Driving Accuracy. To give context on the approximately 9 stroke difference, a tournament can be won or lost in one stroke, and a 5-10 stroke difference can often be the difference between an amateur golfer and a professional golfer

## Importing, Formatting and Cleaning the Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
sns.set_palette("GnBu_d")
```

```python
df = pd.read_csv('pga_data_historical.csv') #2010-2019 PGA Tour data, source: https://www.kaggle.com/bradklassen/pga-tour-20102018-data (as of August 2020, he has taken the data down)
```

```python
df.head()
```

```python
df.shape
```

```python
#Unearthing the available variables in the "statistic" column
df.statistic.unique()
```

```python
df.statistic.value_counts()
```

```python
#Selecting the relevant variables from the "statistic" column

#The dependent variable will be the PGA Tour player's "Scoring Average (Actual)"
#The rest are the independent variables

chosen_statistic = ['Greens in Regulation Percentage',
                    'Driving Accuracy Percentage',
                    'Scrambling',
                    'Scoring Average (Actual)']

#Definitions for each variable:
    #Greens in Regulation Percentage: The percent of time a player was able to hit the green in regulation (greens hit in regulation/holes played). Note: A green is considered hit in regulation if any portion of the ball is touching the putting surface after the GIR stroke has been taken. (The GIR stroke is determined by subtracting 2 from par (1st stroke on a par 3, 2nd on a par 4, 3rd on a par 5))
    #Driving Accuracy Percentage: The percentage of time a tee shot comes to rest in the fairway (regardless of club)
    #Scrambling: The percent of time a player misses the green in regulation, but still makes par or better.
    #Scoring Average (Actual) - The average number of strokes per completed round.
```

```python
#Adjusting the DataFrame
df = df[df['statistic'].isin(chosen_statistic)]
df = df[df['variable'] != 'RANK THIS WEEK'] #Multiple variables have the value 'RANK THIS WEEK', so unstacking will create duplicate columns of 'RANK THIS WEEK', causing an error ("Index contains duplicate entries, cannot reshape"). This line eliminates the error
df = df.drop(columns = ['variable'])
df.head()
```

```python
#Unstacking the data in the DataFrame
df = df.set_index(['player_name', 'date', 'tournament','statistic'])['value'].unstack('statistic').reset_index()
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
print(df.dtypes)
```

```python
#Converting data types from objects to float
df['Greens in Regulation Percentage']=df['Greens in Regulation Percentage'].astype('float')
df['Driving Accuracy Percentage']=df['Driving Accuracy Percentage'].astype('float')
df['Scrambling']=df['Scrambling'].astype('float')
df['Scoring Average (Actual)']=df['Scoring Average (Actual)'].astype('float')
```

```python
print(df.dtypes)
```

```python
#Cleaning up the DataFrame - renaming certain columns
df.rename(columns = {'Greens in Regulation Percentage':'GIR',
                     'Driving Accuracy Percentage':'Fairways',
                     'Scrambling':'Scrambling',
                     'Scoring Average (Actual)':'Score'}, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

### Fixing an error in the data
There is at least one GIR value that is above 100% (in the chart above the max = 487.5).  Will drop any rows where the GIR is above 100%:

```python
df=df[df.GIR <= 100]
```

```python
df.describe()
```

## OLS Summary: GIR and Score

```python
GIR_model = sm.OLS.from_formula("Score ~ GIR", data=df)
GIR_results = GIR_model.fit()
print(GIR_results.summary())

#Note:
#P-value is the probability of getting a sample proportion at least this extreme.  Ranges from zero to 100%. 
#NThe P>|t| value in the OLS summary reflects a two-sided P-value, and needs to be divided by 2 to reflect the single sided value.
#P-values quoted in the summary of this notebook reflect the single sided P-value
```

## OLS Summary: Driving Accuracy and Score

```python
fairways_model = sm.OLS.from_formula("Score ~ Fairways", data=df)
fairways_results = fairways_model.fit()
print(fairways_results.summary())
```

## OLS Summary: Scrambling and Score

```python
scrambling_model = sm.OLS.from_formula("Score ~ Scrambling", data=df)
scrambling_results = scrambling_model.fit()
print(scrambling_results.summary())
```

## Visualizing the data: Checking for linearity
The shape of the data is relatively symmetric, so we will assume a normal distribution

```python
fig, axs = plt.subplots(ncols=3, figsize=(20, 4)) #Default: 6.4, 4.8
sns.scatterplot(x='GIR', y='Score', data=df, ax=axs[0], alpha=0.3, edgecolor=None)
sns.scatterplot(x='Fairways', y='Score', data=df, ax=axs[1], alpha=0.3, edgecolor=None)
sns.scatterplot(x='Scrambling', y='Score', data=df, ax=axs[2], alpha=0.3, edgecolor=None)
fig.suptitle('Scatter plots', fontsize=14)
axs[0].set_title('GIR % vs Score')
axs[0].set_xlabel('GIR %')
axs[0].set_ylabel('Score')
axs[1].set_title('Driving Accuracy % vs Score')
axs[1].set_xlabel('Driving Accuracy %')
axs[1].set_ylabel('')
axs[2].set_title('Scrambling % vs Score')
axs[2].set_xlabel('Scrambling %')
axs[2].set_ylabel('')
plt.show()
```

## Visualizing the residuals: Checking for independence of errors and normality of errors

```python
#Residual Plot: Checking that the residuals are 1) normally distributed, 2) symmetrically distributed around zero with a constant variance.

fig, axs = plt.subplots(ncols=3, figsize=(20, 4))
sns.scatterplot(df.GIR, GIR_results.resid, ax=axs[0], alpha=0.3, edgecolor=None) #Alternative plot, same output: sns.residplot(x='GIR', y='Score', data=df)
sns.scatterplot(df.Fairways, fairways_results.resid, ax=axs[1], alpha=0.3, edgecolor=None)
sns.scatterplot(df.Scrambling, scrambling_results.resid, ax=axs[2], alpha=0.3, edgecolor=None)
fig.suptitle('Residual plots', fontsize=14)
axs[0].set_title('GIR % vs Score')
axs[0].set_xlabel('Fitted Values (GIR %)')
axs[0].set_ylabel('Residuals')
axs[1].set_title('Driving Accuracy % vs Score')
axs[1].set_xlabel('Fitted Values (Driving Accuracy %)')
axs[1].set_ylabel('')
axs[2].set_title('Scrambling % vs Score')
axs[2].set_xlabel('Fitted Values (Scrambling %)')
axs[2].set_ylabel('')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(20, 4))
sm.qqplot(GIR_results.resid, fit=True, ax=axs[0], color='#0A4753', alpha=0.3)
sm.qqplot(fairways_results.resid, fit=True, ax=axs[1], color='#0A4753', alpha=0.3)
sm.qqplot(scrambling_results.resid, fit=True, ax=axs[2], color='#0A4753', alpha=0.3)
fig.suptitle('Normal Probability Plots (Q-Q Plots)', fontsize=14)
axs[0].set_title('GIR % vs Score')
axs[0].set_xlabel('Residual')
axs[0].set_ylabel('Sample Quantiles')
axs[1].set_title('Driving Accuracy % vs Score')
axs[1].set_xlabel('Residual')
axs[1].set_ylabel('')
axs[2].set_title('Scrambling % vs Score')
axs[2].set_xlabel('Residual')
axs[2].set_ylabel('')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(20, 4))
GIR_results.resid.hist(ax=axs[0])
fairways_results.resid.hist(ax=axs[1])
scrambling_results.resid.hist(ax=axs[2])
fig.suptitle('Histogram of Residuals', fontsize=14)
axs[0].set_title('GIR % vs Score')
axs[0].set_xlabel('Residual')
axs[0].set_ylabel('Frequency')
axs[1].set_title('Driving Accuracy % vs Score')
axs[1].set_xlabel('Residual')
axs[2].set_title('Scrambling % vs Score')
axs[2].set_xlabel('Residual')
plt.show()
```
